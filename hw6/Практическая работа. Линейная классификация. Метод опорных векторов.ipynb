### 3. Постройте baseline-модель:
- разбейте TrainData на тренировочные (Train) и тестовые данные (Test); 
- обучите KNN, LogisticRegression и SVC с параметрами по умолчанию на тренировочных данных (Train);
- примените модели на тестовых данных (Test) и вычислите значение метрики f1.

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split

X = Xtrain.drop('target', axis=1)
y = Xtrain['target']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
knn = KNeighborsClassifier()
knn.fit(Xtrain, ytrain)

pred_train = knn.predict(Xtrain)
pred_test = knn.predict(Xtest)
```

```python
print("f score train = ", f1_score(ytrain, pred_train))
print("f score test = ", f1_score(ytest, pred_test))
```

```python
lr = LogisticRegression()
lr.fit(Xtrain, ytrain)

pred_train_lr = lr.predict(Xtrain)
pred_test_lr = lr.predict(Xtest)
```

```python
print("f score train = ", f1_score(ytrain, pred_train_lr))
print("f score test = ", f1_score(ytest, pred_test_lr))
```

```python
svc = SVC()
svc.fit(Xtrain, ytrain)

pred_train_svc = svc.predict(Xtrain)
pred_test_svc = svc.predict(Xtest)
```

```python
print("f score train = ", f1_score(ytrain, pred_train_svc))
print("f score test = ", f1_score(ytest, pred_test_svc))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
   'n_neighbors': list(range(4, 11)),
   'weights': ['uniform', 'distance'],
   'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
}
knn_g = KNeighborsClassifier()

grid_search_knn = GridSearchCV(
   estimator=knn_g,
   param_grid=param_grid,
   scoring='f1',
   n_jobs=-1,
   verbose=True,
)

grid_search_knn.fit(Xtrain, ytrain)

print(grid_search_knn.best_params_)

print(grid_search_knn.best_score_)
```

```python
import numpy as np
param_grid = {
   'C': list(np.arange(0.05, 10.1, 0.1)),
   'class_weight': [None, 'balanced'],
   'solver': ['sag', 'saga'],
   'max_iter': list(range(100, 5001, 100))
}
lr_g = LogisticRegression()

grid_search_lr = GridSearchCV(
   estimator=lr_g,
   param_grid=param_grid,
   scoring='f1',
)

grid_search_lr.fit(Xtrain, ytrain)

print(grid_search_lr.best_params_)

print(grid_search_lr.best_score_)
```
